```python
# Import the pathlib and csv library
from pathlib import Path
import csv

# Set the file path

csv_path = Path('Resources/budget_data.csv')

# Initialize list of records

line_num = 0
monthly_pnl = []
dates = []
change_pnl= []
net_total_pnl = 0
max_profit_increase = 0
max_loss_decrease = 0


sum = 0

# Open the csv file as an object
with open(csv_path, 'r') as csv_file:
   
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csv_reader = csv.reader(csv_file, delimiter=',')
    
     # Read the header row first
    csv_header = next(csv_reader)
   
    
# Read each row of data after the header
    for row in csv_reader:
        line_num += 1
        monthly_pnl.append(int(row[1]))
        net_total_pnl += int(row[1])
        dates.append(row[0])
        
# Iterate through monthly_pnl to Calculate Monthly Change in Profits/Losses
        
for i in range(1, len(monthly_pnl)):
    x = monthly_pnl[i] - monthly_pnl[i - 1]
    change_pnl.append(int(x))
    
#Calculate Average Profit/Loss Change

for i in range(0, len(change_pnl)):
    sum += change_pnl[i]
    avg_change_pnl = round((sum / (len(change_pnl))), 2)

#Determine Max Increase in Profit/Max Decrease in Losses
for monthlypnl in monthly_pnl:
    
    if max_loss_decrease == 0:
        max_profit_increase == monthlypnl
        max_loss_decrease == monthlypnl
    if monthlypnl > max_profit_increase:
        max_profit_increase = monthlypnl
    elif monthlypnl < max_loss_decrease:
        max_loss_decrease = monthlypnl

#Locate Index Values of Greatest Changes in Profit/Losses
max_profit_increase_index = monthly_pnl.index(max_profit_increase)
max_loss_decrease_index = monthly_pnl.index(max_loss_decrease)

max_profit_increase_date = dates[max_profit_increase_index]
max_loss_decrease_date = dates[max_loss_decrease_index]

print(f"Financial Analysis")
print(f"----------------------------")
print(f"Total Months: {line_num} ")
print(f"Total: ${net_total_pnl}")
print(f"Average Change: ${avg_change_pnl}")
print(f"Greatest Increase in Profits: ${max_profit_increase} on {max_profit_increase_date}")
print(f"Greatest Decrease in Profits: ${max_loss_decrease} on {max_loss_decrease_date}")

output_path = Path('Financial_Analysis.txt')

with open(output_path, 'w') as file:
    file.write(f"Financial Analysis\n")
    file.write(f"----------------------------\n")
    file.write(f"Total Months: {line_num} \n")
    file.write(f"Total: ${net_total_pnl}\n")
    file.write(f"Average Change: ${avg_change_pnl}\n")
    file.write(f"Greatest Increase in Profits: {max_profit_increase_date} (${max_profit_increase}) \n")
    file.write(f"Greatest Decrease in Profits: {max_loss_decrease_date} (${max_loss_decrease}) \n")
    
```
